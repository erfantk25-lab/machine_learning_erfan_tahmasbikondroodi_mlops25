```python
import pandas as pd

df = pd.read_csv("advertising.csv", index_col=0)
df
```

```python
# 200 samples
# 3 feauters
# 1 lable
df.shape
```

```python
x, y = df.drop("sales", axis = 1), df["sales"]

x.head()
```

```python
y .head()
```

## Train|test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(

    x, y, test_size=0.33, random_state=42
)

X_train
```

```python
X_test.shape
```

```python
y_train.shape, y_test.shape
```

## Featrue scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)

scaler
```

```python
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)

scaled_X_train.shape, scaled_X_test.shape
```

```python
scaled_X_train.min(), scaled_X_train.max()
```

```python
scaled_X_test.min(), scaled_X_test.max()
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model
```

```python
model.fit(scaled_X_train, y_train)
model.coef_
```

```python
model.intercept_
```

## Prediction

```python
test_sample_feautures = scaled_X_test[0].reshape(1,-1)
test_sample_target = y_test.values[0]
test_sample_feautures, test_sample_target
```

```python
test_sample_feautures.shape
```

```python
model.predict(test_sample_feautures)
```

```python
test_sample_target
```

## Prediction on test data


```python
y_pred = model.predict(scaled_X_test)
y_pred.shape
```

```python
y_test.shape
```

```python
y_test[:5].values
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np 

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"{mae = :.2f}")
print(f"{mse = :.2f}")
print(f"{rmse = :.2f}")
```
